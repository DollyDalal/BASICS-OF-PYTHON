#in slicing the 1st index is inclusive, and the last index is exclusive, it goes till only that last term -1
example-[4:7]
4th index(0,1,2,3rd) and will be printed till 7-1th index

```python
#slicing
thislist=["apple","banana","cherry","orange","kiwi","melon","mango"]
print(thislist[2:6])
```

```python
#slicing
thislist=["apple","banana","cherry","orange","kiwi","melon","mango"]
print(thislist[:6])
print(thislist[2:])
```

```python
#slicing in reverse
thislist=["apple","banana","cherry","orange","kiwi","melon","mango"]
print(thislist[-3:-1])
print(thislist[-3:])
```
